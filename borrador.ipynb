```python
tupla = ('enero', 'febrero', 'marzo', 'abril', 'mayo', 'junio', 'julio', 'agosto', 'septiembre', 'octubre', 'noviembre', 'diciembre')
matriz = [] # como en gastos.csv
diccionario = {} # como en modelo-diccionario.json

def cargarDatos(matriz, archivo):
    with open(archivo) as data:
        primeraLinea = True
        for line in data:
            gastos = line.strip('\n').split(',')
            if not primeraLinea:
                gastos[1] = round(float(gastos[1]), 2) # redondea el importe
            else:
                primeraLinea = False
            gastos[2] = gastos[2].strip() # elimina espacios en blanco
            matriz.append(gastos)
        matriz.pop(0) # elimina cabecera

cargarDatos(matriz, 'gastos.csv')

for line in matriz:
    print(line)
```

```python
''' crear diccionario usando matriz de gastos.csv. 
    el diccionario tendra el formato:
    {
        'enero': {
            'alimentacin': [100, 200, 150],
            'transporte': [50, 80],
            'entretenimiento': [30]
        },
        'febrero': {
            'alimentacin': [100, 200, 150],
            'transporte': [50, 80],
            'entretenimiento': [30]
        },
        'marzo': {
            'alimentacin': [100, 200, 150],
            'transporte': [50, 80],
            'entretenimiento': [30]
        }
    }'''

def crearDiccionario(matriz, diccionario):
    for linea in matriz:
        # linea[0] es de la forma YYYY-MM-DD 
        mes = tupla[int(linea[0].split('-')[1]) - 1]
        concepto = linea[2]
        importe = linea[1]
        if mes not in diccionario:
            diccionario[mes] = {}
        if concepto not in diccionario[mes]:
            diccionario[mes][concepto] = []
        diccionario[mes][concepto].append(importe)

    return diccionario

diccionario = {}
crearDiccionario(matriz, diccionario)

for mes in diccionario:
    print(mes)
    for concepto in diccionario[mes]:
        print(f'\t{concepto}: {diccionario[mes][concepto]}')


```

```python
# obtener de diccionario el total de gastos de un mes
def totalGastosMes(diccionario, mes):
    total = 0
    for categoria in diccionario[mes]:
        total += sum(diccionario[mes][categoria]) # suma los importes de cada categoria
    return total

print(totalGastosMes(diccionario, 'enero'))
```

```python
def mostrarGastos(diccionario, mes):
    print(f'Gastos de {mes}:')
    # calcular sub totales por categorias 
    for categoria in diccionario[mes]:
        total = sum(diccionario[mes][categoria])
        print(f'\t{categoria}: {total}')

mostrarGastos(diccionario, 'enero')
```

```python
# obtener total de gastos por categoria de todo el diccionario
def totalGastosPorCategoria(diccionario):
    total = {}
    for mes in diccionario:
        for categoria in diccionario[mes]:
            if categoria not in total:
                total[categoria] = 0
            total[categoria] += sum(diccionario[mes][categoria])
    return total

total = totalGastosPorCategoria(diccionario)

for k,v in total.items():
    print(f'{k}: ${v}')
    
```

```python
# obtener subtotales por mes 
def totalGastosPorMes(diccionario):
    for mes in tupla:
        print(f'{mes}: ${totalGastosMes(diccionario, mes)}')

totalGastosPorMes(diccionario)
```

```python
# obtener lista de categorias, eliminando duplicados
def listaDeCategorias(diccionario):
    categoria = [x[2] for x in matriz]
    return list(set(categoria))

listaDeCategorias(diccionario)
```

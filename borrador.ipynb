```python
tupla = ('enero', 'febrero', 'marzo', 'abril', 'mayo', 'junio', 'julio', 'agosto', 'septiembre', 'octubre', 'noviembre', 'diciembre')
matriz = [] # como en gastos.csv
diccionario = {} # como en modelo-diccionario.json
```

```python
matriz = []
def cargarDatosConId(matriz, archivo):
    with open(archivo) as data:
        primeraLinea = True
        for line in data:
            gastos = line.strip('\n').split(',')
            if not primeraLinea:
                gastos[0] = int(gastos[0]) # convierte el id a entero
                gastos[2] = round(float(gastos[2]), 2) # redondea el importe
            else:
                primeraLinea = False
            gastos[3] = gastos[3].strip() # elimina espacios en blanco
            matriz.append(gastos)
        matriz.pop(0) # elimina cabecera

cargarDatosConId(matriz, 'gastos-id.csv')

for line in matriz[:10]:
    print(line)
```

```python
def crearDiccionarioId(tupla, matriz, diccionario):
    for linea in matriz:
        # linea[0] es de la forma YYYY-MM-DD 
        mes = tupla[int(linea[1].split('-')[1]) - 1]
        concepto = linea[3]
        importe = linea[2]
        if mes not in diccionario:
            diccionario[mes] = {}
        if concepto not in diccionario[mes]:
            diccionario[mes][concepto] = []
        diccionario[mes][concepto].append(importe)

    

diccionario = {}
crearDiccionarioId(tupla, matriz, diccionario)

for mes in diccionario:
    print(mes)
    for concepto in diccionario[mes]:
        print(f'\t{concepto}: {diccionario[mes][concepto]}')

```

```python
# obtener de diccionario el total de gastos de un mes
def totalGastosMes(diccionario, mes):
    total = 0
    for categoria in diccionario[mes]:
        total += sum(diccionario[mes][categoria]) # suma los importes de cada categoria
    return total

print(totalGastosMes(diccionario, 'enero'))
```

```python
def mostrarGastos(diccionario, mes):
    print(f'Gastos de {mes}:')
    # calcular sub totales por categorias 
    for categoria in diccionario[mes]:
        total = sum(diccionario[mes][categoria])
        print(f'\t{categoria}: {total}')

mostrarGastos(diccionario, 'enero')
```

```python
# obtener total de gastos por categoria de todo el diccionario
def totalGastosPorCategoria(diccionario):
    total = {}
    for mes in diccionario:
        for categoria in diccionario[mes]:
            if categoria not in total:
                total[categoria] = 0
            total[categoria] += sum(diccionario[mes][categoria])
    return total

total = totalGastosPorCategoria(diccionario)

for k,v in total.items():
    print(f'{k}: ${v}')
    
```

```python
# obtener subtotales por mes 
def totalGastosPorMes(diccionario):
    for mes in tupla:
        print(f'{mes}: ${totalGastosMes(diccionario, mes)}')

totalGastosPorMes(diccionario)
```

```python
# obtener lista de categorias, eliminando duplicados
def listaDeCategorias(matriz):
    categoria = [x[3] for x in matriz]
    return list(set(categoria))

listaDeCategorias(matriz)
```

```python
def cargarGasto(matriz):
    
    print(f'Lista de categorias: {listaDeCategorias(matriz)}')
    id = matriz[-1][0] + 1
    fecha = input('Introduce una fecha YYYY-MM-DD: ')
    importe = float(input('Introduce un importe: '))
    categoria = input('Introduce un categoria: ')
    matriz.append([id, fecha, importe, categoria])
    return matriz[-1]

ultimoGasto = cargarGasto(matriz)

# agregar a diccionario el nuevo gasto
crearDiccionarioId(tupla, [ultimoGasto], diccionario)
```

```python
# eliminar de matriz el gasto con id = 993 
def eliminarGasto(matriz, id):
    encontrado = False
    i=0
    while i < len(matriz) and not encontrado:
        if matriz[i][0] == id:
            del matriz[i]
            encontrado = True
        i += 1
    return encontrado

print(eliminarGasto(matriz, 1001))
print(matriz[-10:])
```

```python
def editarGasto(matriz, id, fecha, importe, categoria):
    encontrado = False
    i=0
    while i < len(matriz) and not encontrado:
        if matriz[i][0] == id:
            fecha = fecha
            importe = importe
            categoria = categoria
            matriz[i] = [id, fecha, importe, categoria]
            encontrado = True
        i += 1
    return encontrado

print(editarGasto(matriz, 1000, '2024-08-21', 101, 'alimentación'))
print(matriz[-10:])
```

```python
def buscarGastoPorId(matriz, id):
    for gasto in matriz:
        if gasto[0] == id:
            return gasto
    return None

print(buscarGastoPorId(matriz, 993))
```

```python
def buscarGastoPorCategoria(matriz, categoria):
    gastos = []
    for gasto in matriz:
        if gasto[3] == categoria:
            gastos.append(gasto)
    return gastos

lista = buscarGastoPorCategoria(matriz, 'otros')
for gasto in lista[-10:]:
    print(gasto)
```

```python
def buscarGastoPorFecha(matriz, fecha):
    gastos = []
    for gasto in matriz:
        if gasto[1] == fecha:
            gastos.append(gasto)
    return gastos

lista = buscarGastoPorFecha(matriz, '2024-08-21')
for gasto in lista:
    print(gasto)
    
```

```python
def buscarGastosPorRangoImporte(matriz, minimo, maximo):
    gastos = []
    for gasto in matriz:
        if gasto[2] >= minimo and gasto[2] <= maximo:
            gastos.append(gasto)
    return gastos

lista = buscarGastosPorRangoImporte(matriz, 700, 800)
for gasto in lista[-10:]:
    print(gasto)

    
```
